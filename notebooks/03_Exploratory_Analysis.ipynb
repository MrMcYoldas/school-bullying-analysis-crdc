# 📊 Exploratory Data Analysis (EDA)

## School Bullying Analysis using the CRDC 2021–22 Dataset

---

## Purpose

Before importing the cleaned dataset into PostgreSQL, this notebook explores the structure, quality, and characteristics of the dataset.

The objective is to understand the available variables, identify interesting patterns, and formulate analytical questions that will later be answered using SQL and visualized in Tableau.

This notebook serves as the bridge between the ETL pipeline and the analytical phase of the project.

---

# 🎯 Objectives

By the end of this notebook we aim to:

- Understand the overall structure of the cleaned dataset
- Verify data quality after the ETL process
- Explore distributions of bullying allegations
- Identify patterns across states and school districts
- Explore demographic breakdowns
- Detect potential outliers or unusual observations
- Generate hypotheses for further SQL analysis
- Define visualizations for the final Tableau dashboard
