```python
for file in [
    "SCH/Harassment and Bullying.csv",
    "SCH/Suspensions.csv",
    "SCH/Expulsions.csv",
    "SCH/Referrals and Arrests.csv"
]:
    with zipfile.ZipFile(zip_path, "r") as z:
        with z.open(file) as f:
            temp = pd.read_csv(f, nrows=1)

    print("\n")
    print("="*80)
    print(file)
    print("="*80)

    for col in temp.columns:
        if "DISC" in col or "EXP" in col or "ARREST" in col or "REF" in col or "SUSP" in col:
            print(col)
```

```python
for file in [
    "SCH/Suspensions.csv",
    "SCH/Expulsions.csv",
    "SCH/Referrals and Arrests.csv"
]:
    with zipfile.ZipFile(zip_path, "r") as z:
        with z.open(file) as f:
            temp = pd.read_csv(f, nrows=1)

    print("\n" + "="*80)
    print(file)
    print("="*80)

    for col in temp.columns:
        if any(x in col.upper() for x in ["HB", "BULL", "HARASS"]):
            print(col)
```
